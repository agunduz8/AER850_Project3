```python
# Part 1
from google.colab import drive
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np

drive.mount('/content/drive')

image_path = "/content/drive/My Drive/project3/motherboard_image.JPEG"

img_real = cv2.imread(image_path, cv2.IMREAD_COLOR) #Load it in color mode
img_real = cv2.rotate(img_real, cv2.ROTATE_90_CLOCKWISE) # Rotate 90 degrees clockwise
img = cv2.imread(image_path, cv2.IMREAD_COLOR)
img = cv2.GaussianBlur(img, (47, 47), 4) # Apply Gaussian Blur to smooth the image and reduce noise
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) # Convert it to grayscale
img_gray = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 55,7) # Adaptive thresholding to create a binary image
img_gray = cv2.rotate(img_gray, cv2.ROTATE_90_CLOCKWISE)

edges = cv2.Canny(img_gray, 50, 300) # Canny edge detector to extract edges
edges = cv2.dilate(edges,None, iterations = 10) # Dilate the edges to enhance

#Edge Detection Image
plt.figure(figsize=(8, 8))
plt.imshow(edges, cmap="gray")
plt.title("Edge Detection")
plt.axis("off")
plt.show()

contours, cat = cv2.findContours(edges,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE) # Find contours from the edge-detected image
mask = np.zeros_like(img_real) # Create a blank mask
cv2.drawContours(image=mask, contours=[max(contours, key = cv2.contourArea)], contourIdx=-1, color=(255, 255, 255), thickness= cv2.FILLED) # Draw largest contour onto the mask and use white to fill
masked_img =cv2.bitwise_and(mask, img_real) # Perform bitwise and operation to apply mask
cv2.imwrite("/content/drive/My Drive/project3/masked_motherboard.jpg", masked_img)

#Masked Image
plt.imshow(cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB))
plt.title("Masked Image")
plt.axis("off")
plt.show()

# Part 2
from ultralytics import YOLO

path_data = "/content/drive/MyDrive/project3/data/data.yaml"

# Load the YOLOv8 nano model
model = YOLO("yolov8n.pt")

# Train the model
model.train(
    data=path_data,
    epochs=100,
    batch=6,
    imgsz=1088,
    name="pcb_detector",
    project="/content/drive/MyDrive/project3/results"
)
```
